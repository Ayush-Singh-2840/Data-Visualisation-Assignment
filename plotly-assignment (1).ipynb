Question.1 - Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)


```python
#Answer 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set seed for reproducibility
np.random.seed(30)

# Generate synthetic data
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a figure
fig = plt.figure(figsize=(12, 8))

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')

# Labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot of Random Data')

# Show plot
plt.show()
```

```python
#Answer 

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Violin Plot for Student Grades
# Set seed for reproducibility
np.random.seed(15)

# Generate synthetic data for student grades
data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data_grades)

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=df_grades, palette='viridis')
plt.title('Distribution of Scores Across Different Grade Categories')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.show()

# Heatmap for Sales Data
# Set seed for reproducibility
np.random.seed(20)

# Generate synthetic sales data
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)

# Pivot the data to create a matrix for the heatmap
pivot_table = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=0.5)
plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

Question.4 - Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(x2 + y2 )

x = np.linspace(-5, 5, 100) 
y = np.linspace(-5, 5, 100) 
x, y = np.meshgrid(x, y) 
z = np.sin(np.sqrt(x**2 + y**2))  
data = { 
    'X': x.flatten(), 
    'Y': y.flatten(), 
    'Z': z.flatten() 
} 
df = pd.DataFrame(data)

```python
#Answer 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data for DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Show plot
plt.show()
```

Question.5 - Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

np.random.seed(25) 
data = { 
    'Country': ['USA', 'Canada', 'UK',
'Germany', 'France'], 
    'Population':
np.random.randint(100, 1000, 5), 
    'GDP': np.random.randint(500, 2000,
5) 
} 
df = pd.DataFrame(data)

```python
#Answer

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(25)

# Generate synthetic data
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(
    df['GDP'], 
    df['Population'], 
    s=df['Population'] * 10,  # Bubble size proportional to population
    alpha=0.6,  # Transparency
    c='blue',  # Bubble color
    edgecolors='w',  # Edge color of bubbles
    linewidth=0.5
)

# Annotate each bubble with the country name
for i, country in enumerate(df['Country']):
    plt.text(
        df['GDP'][i], 
        df['Population'][i], 
        country, 
        fontsize=12, 
        ha='right'
    )

# Labels and title
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart of Countries by Population and GDP')

# Show plot
plt.show()
```
